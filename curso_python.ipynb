Faça um programa que peça ao usuário para digitar um número inteiro,
informe se este número é par ou impar. Caso o usuário não digite um número inteiro, informe que não é um número inteiro

```python
def det_int(numero_inteiro):
  if numero_inteiro.isdigit():
    numero_inteiro = int(numero_inteiro)
    if numero_inteiro % 2 == 0:
       return print('Esse número é par')
    else:
       return print('Esse número é impar') 
  
  else:
    return print('Isso não é um número.')
```

```python
numero_inteiro = input('Digite um número inteiro: ')
det_int(numero_inteiro)
```

Formatando valores com modificadores


*   :s - texto (strings)
*   :d - inteiros (int)
* :f - Númeos de ponto flutuante (float)
* :.(numero)f - Quantidade de casas decimais (float)
* :(CARACTERE)(> OU < OU ^)(QUANTIDADE)(TIPO - s,d ou f)












```python
num_1 = 2
```

```python
num_2 = 3
```

```python
divisão = num_1/num_2
```

```python
print('{:.12f}'.format(divisão))
```

```python
print('{:0^11}'.format(num_1))
```

```python
nome = 'Gabriel'
sobrenome = 'Eninger'
nome_formatado = '{0:#^10}{1:#^9}'.format(nome,sobrenome)

print(nome_formatado)
```
